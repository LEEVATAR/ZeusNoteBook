```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.datasets import load_wine
from tensorflow.keras.optimizers import Adam

# 경고 메시지 숨기기
import warnings
warnings.filterwarnings(action='ignore')

# 데이터 불러오기
wine = load_wine()

# 입력과 출력 변수로 나누기
x = wine.data
y = wine.target

# 클래스의 개수 출력
print('클래스 : ', np.unique(y))

# 문자열을 숫자로 변환
e = LabelEncoder()
e.fit(y)
y = e.transform(y)

# 원-핫 인코딩
y_encoded = to_categorical(y)

# 데이터셋 분리
x_train, x_test, y_train_encoded, y_test_encoded = train_test_split(x, y_encoded, test_size=0.3, random_state=91)
_, _, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=91)

# 딥러닝 모델 구성하기
model = Sequential()
model.add(Dense(128, input_dim=x_train.shape[1], activation='relu'))
model.add(Dense(256))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(len(np.unique(y)), activation='softmax'))

# 딥러닝 모델 컴파일
model.compile(loss='categorical_crossentropy',optimizer=Adam(lr=0.001), metrics=['accuracy'])

# 딥러닝 모델 학습
model.fit(x_train, y_train_encoded, validation_data=(x_test, y_test_encoded), epochs=333, batch_size=991)

# 딥러닝 모델 평가
loss, accuracy = model.evaluate(x_test, y_test_encoded)
print('Loss: ', loss)
print('Accuracy: ', accuracy)

# 결과 확인
print(submission)

# 결과를 CSV 파일로 저장
submission.to_csv('submission.csv', index=False)
```
